```python
# remove extra characters 
temp=[re.findall("[a-zA-Z0-9_ @.-""]+",contact_info_df_clean.iloc[i,0]) for i in list(range(len(contact_info_df_clean)))]
temp2=[[j for j in i if j!=' '] for i in temp]
    
temp2
```

```python
clean_df_contact=pd.DataFrame()
for i in list(range(len(temp2))):
    a=pd.DataFrame(temp2[i],  ['contact_id_fake', 'contact_id', 'name_fake', 'name', 'email_fake','email']).transpose()
    clean_df_contact= pd.concat([a,clean_df_contact])
    
clean_df_contact= clean_df_contact[['contact_id','name','email']]
clean_df_contact["contact_id"]=clean_df_contact["contact_id"].str.replace(" ","")
clean_df_contact["contact_id"]=clean_df_contact["contact_id"].astype(np.int64)
clean_df_contact
```
